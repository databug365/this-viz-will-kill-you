```python
import pandas as pd
```

```python
# import transcript data
episode_df = pd.read_parquet("~/grimoire/this-viz-will-kill-you/data/raw/episode_df.parquet")

# validate output
print(episode_df.shape)
display(episode_df["episode_id"].unique())
display(episode_df.head())
```

```python
# 'splode the dataframe so that each word is a row - this is specifically to get the data into a format Tableau will like
word_cloud_df = episode_df.explode("processed_text").rename(columns={"processed_text": "word"}).drop(columns="text")

# group by each word and count the # of occurences
word_cloud_df = word_cloud_df.groupby(["episode_id", "episode_name", "word"]).size().reset_index(name="count")

# sort by by count descending
word_cloud_df = word_cloud_df.sort_values(by=["episode_id","count"], ascending=False)

# set up the value to filter the dataframe
n = 10 # can be adjusted as needed

# filter the dataframe
word_cloud_df = word_cloud_df.loc[word_cloud_df["count"] >= n]

# check output
display(word_cloud_df.head())
```

```python
# get simple metrics for validation
print(word_cloud_df.shape)
display(word_cloud_df["episode_id"].unique())
display(word_cloud_df.head())

# save to drive
word_cloud_df.to_parquet("~/grimoire/this-viz-will-kill-you/data/tidy/word_cloud_df.parquet")
```

```python
# save the word cloud df to csv for Tableau
word_cloud_df.to_csv("~/grimoire/this-viz-will-kill-you/data/publish/word_cloud_df.csv")
```
